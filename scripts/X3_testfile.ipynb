```python
import pandas as pd
```

```python
backtest_df= pd.read_csv('../data/results/backtest_yearly.csv')#

backtest_df.set_index('year', inplace=True)
print(backtest_df)
```

```python
def backtest_diag(df):
    #arithmetic_average_return= df.mean()
    geometric_average_return= ((1 + df).prod() ** (1 / len(df))) - 1
    volatility= df.std()
    max_drawdown= df.min()
    print(df.index[0],df.index[1])



    dg=pd.DataFrame({
        'period':str(df.index[0])+'-'+str(df.index[-11]),
        'annual return': geometric_average_return,
        'volatility':volatility,
        'sharpe': geometric_average_return/volatility,
        'max drawdown': max_drawdown
    })
    return dg
```

```python

backtest_diag(backtest_df)
```
