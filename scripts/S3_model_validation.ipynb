## Compare of XGB parameters

```python
XGBModels={}
for j1 in range(1, 3):
    for j2 in range(1, 2):
        for j3 in range(1, 3):
            model=XGB(
                    n_estimators=500*j1,
                    learning_rate=0.01*j2,
                    max_depth=j3,
                    subsample=0.6,
                    colsample_bytree=0.8,
                    reg_lambda=10,
                    reg_alpha=1,
                    random_state=42,
                    )

            name= 'XGB' + str(j1) + '_' + str(j2) + '_' + str(j3)
            XGBModels[name]= model


            print(name+': n_estimators =' + str(500*j1) + ' learning_rate=' + str(0.01*j2) + ', max_depth =' + str(j3))
```

```python
eval_summary_XGB, eval_mean_XGB=run(XGBModels,"1970-01-31","1999-12-31",train_years = 20, valid_years = 1,step_years = 1 ,max_steps = 10,print_years=True)
```

```python
display(eval_summary_XGB)
#display(eval_mean_XGB.loc['average'])
display(eval_mean_XGB.sort_values(by=['average'], ascending=False, axis=1))


eval_stat =eval_summary_XGB.agg(['mean','std','min'])



display(eval_stat)
```

## Comparisson of random forest models

```python
RFModels={}


for n_estimators, max_depth, min_samples_leaf in product(
    [50, 100, 300,500],
    [None, 10],
    [1, 5]
):
    model= RF(
        n_estimators=n_estimators,
        max_depth=max_depth,
        min_samples_split=10,
        min_samples_leaf=min_samples_leaf,
        random_state=42,
        n_jobs=-1
    )
    name= 'RF' + str(n_estimators) + '_' + str(max_depth) + '_' + str(min_samples_leaf)
    RFModels[name]= model
```

```python
eval_summary_RF, eval_mean_RF=run(RFModels,"1970-01-31","1999-12-31",train_years = 20, valid_years = 1,step_years = 1 ,max_steps = 3,print_years=True )
```
