```python
train_start_date="1970-01-31"
test_end_date="1999-12-31"

df= pd.read_csv('../data/raw/prices.csv', index_col='Date', parse_dates=True)

df = df.loc[train_start_date:test_end_date]


df_bench = pd.read_csv('../data/raw/prices_bench.csv', index_col='Date', parse_dates=True)
df_bench = df_bench.loc[train_start_date:test_end_date]

daily_ret = df.pct_change()
daily_ret_bench = df_bench.pct_change()
daily_ret = daily_ret.loc['1970-02-03':]
daily_ret_bench = daily_ret_bench.loc['1970-02-03':]
rolling_window = 252
min_periods = 126

#display(daily_ret_bench)


#display(daily_ret.rolling(window=rolling_window, min_periods=min_periods).cov(daily_ret_bench.loc[:,'^GSPC']))


beta_1y_daily = (
        daily_ret
        .rolling(window=rolling_window, min_periods=min_periods)
        .cov(daily_ret_bench.loc[:,'^GSPC'])
        .div(
            daily_ret_bench.loc[:,'^GSPC'].rolling(window=rolling_window, min_periods=min_periods).var(),
            axis=0
        )
    )
display(beta_1y_daily)
```
